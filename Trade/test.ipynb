```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pytz
```

```python
df = pd.read_csv('C:\\Users\\Wentao\\Desktop\\EA_recherche\\Trade\\sanofi_1min_aggregation.csv')
df['trade_time_paris'] = pd.to_datetime(df['trade_time_paris'])
df.set_index('trade_time_paris', inplace=True)
plt.plot(df.index, df['x_t_signed_volume'])
ax = plt.gca()
paris_tz = pytz.timezone('Europe/Paris')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M', tz=paris_tz))
```
